# 2.Analyzing the Dataset

```python
# Load the dataset
data_path = "/kaggle/input/veremi-extension-data-1-21-gb/mixalldata_clean.csv"
df = pd.read_csv(data_path)

# Display the first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Display dataset information
print("\nDataset Information:")
df.info()

# Display summary statistics
print("\nSummary Statistics:")
display(df.describe())
```

```python
# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())


# Check for NaN values
print("Number of NaN values in each column:")
print(df.isnull().sum())

# Check for infinite values
print("\nNumber of infinite values in each column:")
print(np.isinf(df.select_dtypes(include=[np.number])).sum())

# Identify duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"Number of duplicate rows: {duplicate_rows.shape[0]}")
```

# 2.2 Checking Class distribution

```python
# Convert all column names to lowercase
df.columns = df.columns.str.lower()

print('Completed converting all column names into lower case')

# Display the distribution of the 'class' column
print("\nClass Distribution:")
print(df['class'].value_counts())

# As a pandas DataFrame
unique_counts = df.nunique().reset_index()
unique_counts.columns = ['Column', 'Unique Values']
print(unique_counts)
```

# 2.3 Plotting Class distribution and histograms for numerical features

```python
# Plot class distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='class', data=df, order=df['class'].value_counts().index)
plt.title('Class Distribution')
plt.xlabel('Class Label')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot histograms for numerical features
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_features].hist(figsize=(15, 10), bins=30)
plt.suptitle('Histograms of Numerical Features')
plt.tight_layout()
plt.show()
```

# 2.5 Deleting 0 variance columns

```python
# Identify columns with zero variance (only one unique value)
zero_variance_cols = [col for col in df.columns if df[col].nunique() == 1]
print("Columns with zero variance:", zero_variance_cols)

# Drop those columns from the DataFrame
df.drop(columns=zero_variance_cols, inplace=True)

# Dynamically identify numerical columns that still exist
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns

print('Completed dropping columns with zero variance')
print('Remaining columns:')
print(df.columns.tolist())

# 2. compute variances of only the feature columns (exclude “class”)
variances = df.drop(columns="class").var()

# 3. sort ascending and display
near_zero = variances.sort_values().to_frame(name="variance")
print(near_zero.head(10))
```

# 2.6 Drawing Correlation matrix

```python
# Compute the correlation matrix
correlation_matrix = df[numerical_features].corr()

# Plot the correlation matrix
plt.figure(figsize=(14, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```

# 3.1 Splitting the data to train/val/test and Standerization

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Assuming 'df' is your DataFrame and 'class' is the target column
X = df.drop('class', axis=1)
y = df['class']

# Stratified split to maintain class proportions
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print('Step 1 Done')

# Further split training data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.2, stratify=y_temp, random_state=42
)
print('Step 2 Done')

# Create the scaler
scaler = StandardScaler()

# Fit on training data only
scaler.fit(X_train)

# Transform all sets
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrame to preserve column names (optional)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=X_val.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)

# Check the distribution of scaled features
print("Mean of scaled training features:", np.mean(X_train_scaled, axis=0)[:5])  # First 5 features
print("Std of scaled training features:", np.std(X_train_scaled, axis=0)[:5])    # First 5 features

print('Step 3 (Standardization) Done')
```
